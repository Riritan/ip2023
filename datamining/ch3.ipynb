```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data =np.array([[45,3],[9,12],[15,10],[60,5],[21,15]]) #np쓰는이유 : :,0 데이터를 원하는것만 들고오기 위해서 사용  2차원배열에서  -열 ㅣ행 np array는 ㅣ묶어서 사용가능
#data [0] 인덱싱 - 리스트에,로구분되어진 여러 아이템이 있을때 하나를 들고오는것 data[:]슬라이싱:범위로 들고오는것. data[0]일경우 [45,3]이 불러와짐 [시작:끝] for i in range[시작:끝:증감] 
#data[행,열] :슬라이싱 -> 처음부터 끝까지 [:,0]= 모든행에 0번째열 .data[0,0]advacing indexing data [:,: ]advancing slicing ->adcancing 권고 
#중간고사 나옴 결과얻기위해 어케나올것인가?? 요런거 
data[0] 
#data [0] data[0:1] 인덱싱은 차원이 축소됨 슬라이싱은 차원유지 
```

```python
plt.scatter(x=data[:,0],y=data[:,1]) #같은 행과열의 점그래프를 그려줌 x,y의 상관계수를 파악 
plt.xlabel('X')
plt.xlabel('Y')
plt.xlim(0,65)#limit
plt.ylim(0,20)

#행과 열의수

rows,columns =data.shape

#데이터 라벨

for i in range(rows):        #알고리즘 돌리기전에 데이터분석먼저 
    plt.annotate(i+1,(data[i,0],data[i,1]))
```

```python
scaler = StandardScaler() #표준화 스케일러 객체생성
scaler.fit(data)#표준화 스케일러를 data에 기준을 맞춤
data_std =scaler.transform(data)#각 열별로 표준화함 0을기준으로 양쪽 스케일링 -범위
print(data_std) 
```

```python
plt.scatter(x=data_std[:,0],y=data_std[:,1])
plt.xlabel('X')
plt.xlabel('Y')
plt.xlim(-2,2)#limit
plt.ylim(-2,2)

#행과 열의수

rows,columns =data_std.shape

#데이터 라벨

for i in range(rows):         
    plt.annotate(i+1,(data_std[i,0],data_std[i,1]))
```

```python
#군집화
k=2
kmeans =KMeans(n_clusters=2)
kmeans.fit(data_std[:,(0,1)])
print(kmeans.labels_)
print(kmeans.cluster_centers_)
```
